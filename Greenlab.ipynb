```python
import pandas as pd
import scipy.stats as stats
import sys, os

import statsmodels.api as sm
from statsmodels.formula.api import ols



# A one-way ANOVA is primarily designed to enable the equality testing between three or more means. 
# A two-way ANOVA is designed to assess the interrelationship of two independent variables on a dependent variable




# load data file
dd = pd.read_csv("avg-chrome2v2.csv")
#print("df columns are:")
#print(df.columns)
# reshape the d dataframe suitable for statsmodels package 
dd_melt = pd.melt(dd.reset_index(), id_vars=['index'], value_vars=['C','C++', 'Go', 'Rust'])
# replace column names
dd_melt.columns = ['index', 'treatments', 'value']





# stats f_oneway functions takes the groups as input and returns ANOVA F and p value
fvalue, pvalue = stats.f_oneway(dd['C'], dd['C++'],dd['Go'], dd['Rust'])
print(fvalue, pvalue)
# 17.492810457516338 2.639241146210922e-05

# Ordinary Least Squares (OLS) model
model = ols('value ~ C(treatments)', data=dd_melt).fit()
anova_table = sm.stats.anova_lm(model, typ=2)
print(anova_table)
```
